```python
import pandas
from sklearn.linear_model import LinearRegression
from matplotlib import pyplot
```

```python
url="https://raw.githubusercontent.com/sohamglobal/datasets/refs/heads/main/FitnessCustomers.csv"
df=pandas.read_csv(url)
print(df)
```

```python
X=df[['Weight']]
Y=df[['ExerciseTime']]
```

```python
model=LinearRegression()
model.fit(X,Y)
```

```python
weight_values = [[61], [79]]
predicted_prices = model.predict(weight_values)
print(predicted_prices)
```

```python
for weight,price in zip(weight_values,predicted_prices):
  print(f'Weight: {weight[0]} -> {price[0]}')
```
