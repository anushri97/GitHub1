```python
import pandas
import matplotlib.pyplot as plt
```

```python
url="https://raw.githubusercontent.com/sohamglobal/datasets/refs/heads/main/FitnessCustomers.csv"
df=pandas.read_csv(url)
print(df)
```

```python
plt.scatter(df['Age'], df['ExerciseTime'])
plt.title('Age vs Exercise Time')
plt.xlabel('Age')
plt.ylabel('Exercise Time (minutes)')
plt.grid(True)
plt.show()

```

```python
plt.bar(df['Age'],df['ExerciseTime'],color='skyblue')
plt.title('age vs exercise time')
plt.xlabel('age')
plt.ylabel('exercise time')
plt.show()
```

```python
plt.barh(df['Age'],df['ExerciseTime'],color='purple')
plt.title('age vs exercise time')
plt.xlabel('age')
plt.ylabel('exercise time')
plt.show()
```

```python
gender_avg = df.groupby('Gender')['ExerciseTime'].mean()

plt.pie(gender_avg, labels=gender_avg.index, autopct='%1.1f%%')
plt.title('Average Exercise Time by Gender')
plt.show()

```
