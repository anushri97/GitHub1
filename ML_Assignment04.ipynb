```python
import pandas
from sklearn.cluster import KMeans
from matplotlib import pyplot
```

```python
url="https://raw.githubusercontent.com/sohamglobal/datasets/refs/heads/main/FitnessCustomers.csv"
df=pandas.read_csv(url)
print(df)
```

```python
X=df[['Age', 'Weight', 'Height', 'BMI', 'ExerciseTime', 'CaloriesIntake']]
```

```python
km=KMeans(n_clusters=3,random_state=42)
df['Cluster']=km.fit_predict(X)
```

```python
print('Clustered Data : ')
print(df)
```

```python
import matplotlib.pyplot as plt

plt.scatter(df['Age'], df['Weight'], c=df['Cluster'], cmap='viridis')
plt.xlabel('Age')
plt.ylabel('Weight')
plt.title('K-Means Clustering')
plt.colorbar(label='Cluster')
plt.grid(True)



pyplot.show()
```
